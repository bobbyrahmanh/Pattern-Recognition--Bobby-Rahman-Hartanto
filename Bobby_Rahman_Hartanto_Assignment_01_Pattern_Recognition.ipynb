```python
import math

def preprocess_text(text):
    """
    Fungsi untuk case folding (lowercase) dan menghapus tanda baca.
    """

    text = text.lower()

    punctuations = '''!()-[]{};:'"\\,<>./?@#$%^&*_~'''
    for char in punctuations:
        text = text.replace(char, " ")

    tokens = text.split()
    return tokens

def compute_tf(document_tokens):
    """
    Menghitung Term Frequency (TF) untuk satu dokumen.
    """
    tf_dict = {}
    total_words = len(document_tokens)

    for word in document_tokens:
        tf_dict[word] = tf_dict.get(word, 0) + 1

    for word in tf_dict:
        tf_dict[word] = tf_dict[word] / total_words

    return tf_dict

def compute_idf(corpus_tokens):
    """
    Menghitung Inverse Document Frequency (IDF) untuk seluruh korpus.
    """
    idf_dict = {}
    N = len(corpus_tokens)

    all_words = set(word for doc in corpus_tokens for word in doc)

    for word in all_words:
        doc_count = sum(1 for doc in corpus_tokens if word in doc)

        idf_dict[word] = math.log10(N / float(doc_count))

    return idf_dict

def compute_tfidf(corpus):
    """
    Fungsi untuk menjalankan preprocessing + TF-IDF
    """
    tokenized_docs = [preprocess_text(doc) for doc in corpus]

    tf_list = [compute_tf(doc) for doc in tokenized_docs]
    idf_dict = compute_idf(tokenized_docs)

    tfidf_result = []

    for tf in tf_list:
        tfidf = {}
        for word, tf_value in tf.items():
            tfidf[word] = tf_value * idf_dict[word]
        tfidf_result.append(tfidf)

    return tfidf_result

# ==========================================
# contoh
# ==========================================
if __name__ == "__main__":
    # data teks mentah
    data_teks = [
        "Kandasnya hubungan romantis sering kali memicu tingkat stres emosional yang signifikan.",
        "Proses pemulihan pasca putus cinta membutuhkan waktu serta dukungan sosial yang memadai.",
        "Sekarang waktunya fokus untuk perlahan bangkit dan upgrade diri daripada terus memikirkan masa lalu."
    ]

    hasil_tfidf = compute_tfidf(data_teks)

    for i, doc_tfidf in enumerate(hasil_tfidf):
        print(f"=== Dokumen {i+1} ===")
        sorted_tfidf = dict(sorted(doc_tfidf.items(), key=lambda item: item[1], reverse=True))
        for word, score in sorted_tfidf.items():
            print(f"{word}: {score:.4f}")
        print("\n")
```
